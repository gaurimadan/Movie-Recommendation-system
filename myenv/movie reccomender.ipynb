```python
import pandas as pd
import numpy as np
movies=pd.read_csv("D:\Movie-recommender-system\movies.csv")
credits=pd.read_csv("D:\Movie-recommender-system\credits.csv")
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movies['genres']=movies['genres'].apply(convert)

movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["overview"][0]
```

```python
movies["overview"]=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"]=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tag']=movies['genres']+movies['overview']+movies['keywords']
```

```python
movies.head()
```

```python
new_movies=movies[["id",'title',"tag"]]
```

```python
new_movies.head()
```

```python
new_movies['tag']=new_movies['tag'].apply(lambda x:' '.join(x))
```

```python
new_movies.head()
```

```python
new_movies['tag'][0]
```

```python
new_movies['tag']=new_movies['tag'].apply(lambda X:X.lower())
```

```python
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_movies['tag']).toarray().shape
```

```python
vector=cv.fit_transform(new_movies['tag']).toarray()
```

```python
vector[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer() 
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies['tag']=new_movies['tag'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector).shape
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommender(movie):
    movie_index=new_movies[new_movies["title"]==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```
